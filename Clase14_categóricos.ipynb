```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
tips = sns.load_dataset("tips")
tips.head(2)
```

#### Gráficos categóricos

| Tipo de Gráfico        | Función de Seaborn       | Descripción                                                 |
|------------------------|--------------------------|-------------------------------------------------------------|
| Gráfico de Recuento     | `sns.countplot`          | Muestra el recuento de observaciones en una variable categórica. |
| Gráfico de Tira        | `sns.stripplot`           | Muestra distribuciones individuales de datos categóricos. |
| Gráfico de Enjambre     | `sns.swarmplot`          | Combina el gráfico de tira con puntos que evitan la superposición. |
| Gráfico de Caja        | `sns.boxplot`             | Visualiza la distribución de una variable categórica mediante cuartiles. |
| Gráfico de Violín      | `sns.violinplot`          | Combina un gráfico de caja y un gráfico de KDE para mostrar distribuciones. |
| Gráfico de Puntos      | `sns.pointplot`           | Muestra la tendencia central y el intervalo de confianza de una variable categórica. |


```python
sns.countplot(data=tips,x='day',hue='sex')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.stripplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

### Se pueden hacer gráficos combinados para mejorar la visualización

```python
plt.figure(figsize=(6,6)) # Configura el tamaño de la figura

sns.boxplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True) # Crea un diagrama de caja utilizando Seaborn y muestra los datos de total_bill agrupados por día y género

sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True, palette='dark:0',marker='o') # Crea un gráfico swarm utilizando Seaborn y muestra la distribución de los datos de total_bill agrupados por día y género

plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex',split=True,palette='pastel')
plt.show()
```

```python
sns.violinplot(data=tips,y='total_bill')
```

## Al igual que con displot, se puede usar sns.catplot()

sns.catplot() define el mejor gráfico de acuerdo a nuestros datos.

### Parámetros de sns.catplot

| Parámetro       | Descripción                                                                   |
|-----------------|-------------------------------------------------------------------------------|
| `data`          | Conjunto de datos (DataFrame) para la visualización.                         |
| `x`             | Variable en el eje x para agrupar y categorizar.                              |
| `y`             | Variable en el eje y para representar valores.                                |
| `hue`           | Variable categórica para distinguir elementos por color.                      |
| `col`           | Variable categórica para dividir el gráfico en columnas.                       |
| `row`           | Variable categórica para dividir el gráfico en filas.                          |
| `kind`          | Tipo de gráfico categórico a crear ('strip', 'swarm', 'box', 'violin', etc.). |
| `order`         | Orden de las categorías en el eje x.                                          |
| `col_order`     | Orden de las categorías en la división por columnas.                          |
| `row_order`     | Orden de las categorías en la división por filas.                             |
| `orient`        | Orientación del gráfico ('v' para vertical, 'h' para horizontal).             |
| `height`        | Altura de cada aspecto del gráfico.                                           |
| `aspect`        | Relación de aspecto de cada aspecto del gráfico.                               |
| `palette`       | Paleta de colores para elementos categóricos.                                 |
| `dodge`         | Si los elementos deben separarse en la misma categoría.                        |
| `share{x,y}`    | Compartir ejes entre aspectos (True o False).                                  |
| `legend`        | Mostrar la leyenda (True o False).                                            |
| `legend_out`    | Posición de la leyenda ('True', 'brief' o ubicación).                          |


```python
# Crea un gráfico categórico utilizando Seaborn y muestra un conjunto de diagramas de caja
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True,kind='box',col='time')

plt.show()
```

```python
sns.catplot(data = tips, x="day", y ='total_bill', hue="smoker", col="time", dodge=True)

plt.show()
```
