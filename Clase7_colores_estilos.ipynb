## Matlab

```python
fig, ax = plt.subplots(figsize=(8, 8))   # Creamos una figura con un solo eje y tamaño de 8x8

ax.plot(x, x+1, 'r--')   # Graficamos con línea roja punteada
ax.plot(x, x+2, 'b-')    # Graficamos con línea azul continua
ax.plot(x, x+3, 'y.-')   # Graficamos con línea amarilla de puntos con línea
ax.plot(x, x+4, 'go:')   # Graficamos con línea verde circular con línea punteada
```

## Pyplot

ax.plot(atributos)

| Atributo     | Uso                                                  |
|--------------|------------------------------------------------------|
| `linewidth`  | Grosor de la línea.                                 |
| `color`      | Color de la línea. Puede ser nombre o código RGB.   |
| `linestyle`  | Estilo de la línea ('-', ':', '--', etc.).           |
| `marker`     | Marcador de puntos en la línea ('o', 's', 'x', etc.). |
| `markersize` | Tamaño de los marcadores.                           |
| `markeredgecolor` | Color del borde de los marcadores.           |
| `markerfacecolor` | Color del relleno de los marcadores.        |
| `alpha`      | Transparencia de la línea (entre 0 y 1).           |
| `label`      | Etiqueta para la leyenda del gráfico.               |


```python
fig, ax = plt.subplots(figsize=(8, 8))   # Creamos una figura con un solo eje y tamaño de 8x8

ax.plot(x, x+1, color='green', alpha=0.8, linewidth=5, linestyle='-', marker='o', markersize=10, markerfacecolor='#CCFFCC')
# Graficamos x vs x+1 con línea verde, alpha de 0.8 (transparencia), ancho de línea 5, estilo de línea continua ('-'), marcador circular ('o') de tamaño 10 y relleno del marcador con color '#CCFFCC'

ax.plot(x, x+2, color='blue', linewidth=8, linestyle='--', marker='8')
# Graficamos x vs x+2 con línea azul, ancho de línea 8, estilo de línea punteada ('--'), marcador '8'

ax.plot(x, x+3, color='#66FF66', linewidth=3, linestyle='dashed', marker='v')
# Graficamos x vs x+3 con línea de color personalizado (#66FF66), ancho de línea 3, estilo de línea discontinua, marcador de flecha ('v')

ax.plot(x, x+4, color='#CCFFCC', linewidth=2, linestyle=':', marker='P')
# Graficamos x vs x+4 con línea de color personalizado (#CCFFCC), ancho de línea 2, estilo de línea punteada, marcador en forma de pentágono ('P')

ax.set_xlabel('Eje X')   # Etiqueta del eje x
ax.set_ylabel('Eje Y')   # Etiqueta del eje y
ax.set_title('Gráficos con personalización')   # Título del gráfico
ax.legend(['Línea 1', 'Línea 2', 'Línea 3', 'Línea 4'])   # Agregamos la leyenda
```
