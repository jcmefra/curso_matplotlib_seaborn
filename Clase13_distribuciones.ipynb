```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
tips = sns.load_dataset("tips")

tips.head()
```

STAT:
    count: show the number of observations in each bin

    frequency: show the number of observations divided by the bin width

    probability: or proportion: normalize such that bar heights sum to 1

    percent: normalize such that bar heights sum to 100

    density: normalize such that the total area of the histogram equals 1

### Crear distribuciones

| Tipo de Gráfico              | Función de Seaborn       | Descripción                                                |
|------------------------------|--------------------------|------------------------------------------------------------|
| Gráfico de Histograma        | `sns.histplot`           | Muestra la distribución de una variable numérica.          |
| Gráfico de KDE               | `sns.kdeplot`            | Estima y visualiza la función de densidad de kernel.       |
| Gráfico de ECDF              | `sns.ecdfplot`           | Muestra la función de distribución acumulativa empírica.   |



```python
# Crea un histograma utilizando Seaborn y muestra distribuciones de propinas

sns.histplot(data=tips, x='tip',bins=15,cumulative=False,hue='sex',stat='count',multiple='dodge')
plt.show()
```

bw_adjust -- smooth

```python
sns.kdeplot(data=tips,x='tip',hue='sex',cumulative=False, fill=True,bw_adjust=1)
plt.show()
```

```python
sns.ecdfplot(data=tips,x='tip',hue='sex',stat='count')
plt.show()
```

```python
sns.displot(data=tips, x='tip', hue='smoker', kind='kde', multiple='stack')
plt.show()
```
