```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
tips = sns.load_dataset("tips")
tips.head(2)
```

## Gráficos compuestos

| Tipo de Gráfico      | Función de Seaborn        | Descripción                                                                                        |
|----------------------|---------------------------|----------------------------------------------------------------------------------------------------|
| Gráfico de Conjunto   | `sns.jointplot`           | Muestra la relación entre dos variables numéricas con histogramas y un gráfico de dispersión.   |
| Gráfico de Pares     | `sns.pairplot`            | Muestra las relaciones entre múltiples variables numéricas en un conjunto de gráficos de dispersión. |

```python
sns.jointplot(data=tips, x='total_bill', y= 'tip', hue = 'sex')
plt.show()
```

### Se pueden agregar más elementos para hacer un mejor análisis

| Parámetro       | Descripción                                                                                            |
|-----------------|--------------------------------------------------------------------------------------------------------|
| `data`          | Conjunto de datos (DataFrame) para la visualización.                                                  |
| `x`             | Variable en el eje x para los datos.                                                                  |
| `y`             | Variable en el eje y para los datos.                                                                  |
| `hue`           | Variable categórica para distinguir elementos por color.                                              |
| `palette`       | Paleta de colores para elementos categóricos.                                                        |
| `style`         | Variable categórica para distinguir elementos por estilo (por ejemplo, estilo de línea).            |
| `markers`       | Variable categórica para distinguir elementos por marcador.                                           |
| `col`           | Variable categórica para dividir el gráfico en columnas (facetas).                                    |
| `row`           | Variable categórica para dividir el gráfico en filas (facetas).                                        |
| `kind`          | Tipo específico de gráfico (por ejemplo, `'scatter'`, `'kde'`, `'reg'`, `'hex'`).                  |
| `height` y `aspect`| Alto y proporción de aspecto de los gráficos en `pairplot`.                                           |
| `diag_kind`     | Tipo de gráfico para la diagonal en `pairplot` (`'hist'`, `'kde'`).                                  |
| `corner`        | Si `True`, crea gráficos en estilo de esquina en `pairplot`.                                          |
| `vars`          | Variables específicas para incluir en `pairplot`.                                                     |
| `corner_kws`    | Argumentos pasados a la función de gráficos de esquina en `pairplot`.                                 |
| `plot_kws`      | Argumentos pasados a la función de gráficos principales en `pairplot`.                                |
| `marginal_kws`  | Argumentos pasados a las funciones de gráficos marginales en `jointplot` (`kde_kws`, `hist_kws`).   |
| `marginal_ticks`  | Muestra una pequeña tabla para la gráfica exterior   |
| `height`        | Alto del gráfico en `jointplot`.                                                                      |
| `ratio`         | Relación entre el tamaño de la figura y el tamaño del gráfico de dispersión en `jointplot`.          |
| `space`         | Espacio adicional entre el gráfico de dispersión y los histogramas en `jointplot`.                   |


```python
sns.jointplot(data=tips, x='total_bill', y= 'tip', hue='sex', kind= 'hist', marginal_ticks=True, #muestra una pequeña tabla para la gráfica exterior
marginal_kws=dict(bins= 25, fill = True, multiple= 'dodge')); #marginal_kws se le pueden agregar diferentes parámetros
plt.show()
```

### Para sns.pairplot es similar

**Pairplot correlaciona todas las variables numéricas que hay en el dataset**

| Parámetro           | Descripción                                                                                           |
|---------------------|-------------------------------------------------------------------------------------------------------|
| `data`              | Conjunto de datos (DataFrame) para la visualización.                                                 |
| `vars`              | Variables específicas para incluir en el gráfico.                                                    |
| `x_vars`, `y_vars`  | Variables específicas para los ejes x e y.                                                           |
| `hue`               | Variable categórica para distinguir elementos por color.                                             |
| `palette`           | Paleta de colores para elementos categóricos.                                                       |
| `diag_kind`         | Tipo de gráfico para la diagonal (`'hist'`, `'kde'`).                                               |
| `markers`           | Variable categórica para distinguir elementos por marcador.                                          |
| `height` y `aspect` | Alto y proporción de aspecto de los gráficos.                                                       |
| `corner`            | Si `True`, crea gráficos en estilo de esquina.                                                       |
| `corner_kws`        | Argumentos pasados a la función de gráficos de esquina.                                              |
| `plot_kws`          | Argumentos pasados a la función de gráficos principales.                                             |
| `diag_kws`          | Argumentos pasados a las funciones de gráficos diagonales (`kde_kws`, `hist_kws`).                  |
| `grid_kws`          | Argumentos pasados al método `PairGrid`.                                                             |
| `hue_order`         | Orden de los niveles de la variable `hue`.                                                           |
| `markers`           | Variables categóricas para distinguir elementos por marcador.                                        |
| `dropna`            | Si `True`, elimina filas con valores faltantes antes de trazar.                                     |
| `plot_kws`          | Argumentos pasados a las funciones de gráficos en la diagonal principal.                            |
| `diag_kws`          | Argumentos pasados a las funciones de gráficos en la diagonal (`kde_kws`, `hist_kws`).               |
| `hue_kws`           | Argumentos pasados a las funciones de gráficos por nivel de `hue`.                                  |
| `corner_kws`        | Argumentos pasados a las funciones de gráficos en las esquinas.                                      |
| `grid_kws`          | Argumentos pasados al método `PairGrid`.                                                             |

```python
sns.pairplot(data=tips,hue='sex',palette='coolwarm',corner=True,kind='scatter')
plt.show()
```
