```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

## Parámetros de Seaborn

| Parámetro       | Descripción                                                                               |
|-----------------|-------------------------------------------------------------------------------------------|
| `data`          | Conjunto de datos (DataFrame) para la visualización.                                     |
| `x`             | Variable en el eje x para los datos.                                                     |
| `y`             | Variable en el eje y para los datos.                                                     |
| `hue`           | Variable categórica para distinguir elementos por color.                                  |
| `palette`       | Paleta de colores para elementos categóricos.                                            |
| `style`         | Variable categórica para distinguir elementos por estilo (por ejemplo, estilo de línea). |
| `size`          | Variable categórica para distinguir elementos por tamaño.                                 |
| `markers`       | Variable categórica para distinguir elementos por marcador.                               |
| `col`           | Variable categórica para dividir el gráfico en columnas (facetas).                        |
| `row`           | Variable categórica para dividir el gráfico en filas (facetas).                            |
| `kind`          | Tipo específico de gráfico (por ejemplo, 'scatter', 'line', 'hist').                     |
| `col_wrap`      | Número de columnas en las facetas.                                                        |
| `aspect`        | Proporción del aspecto de las facetas.                                                    |
| `col_order`     | Orden de las columnas en las facetas.                                                     |
| `row_order`     | Orden de las filas en las facetas.                                                        |
| `legend`        | Mostrar leyenda (generalmente True o False).                                              |
| `xlabel` y `ylabel` | Etiquetas para los ejes x e y respectivamente.                                          |
| `title`         | Título del gráfico.                                                                       |
| `xlim` y `ylim` | Límites para los ejes x e y respectivamente.                                            |
| `xticks` y `yticks` | Ubicaciones de las marcas en los ejes x e y respectivamente.                         |
| `xticklabels` y `yticklabels` | Etiquetas de las marcas en los ejes x e y respectivamente.                         |


```python
# Cargar un dataset de ejemplo
tips = sns.load_dataset('tips')
```

```python
tips
```

Al usar sns.displot, Seaborn me recomienda el mejor gráfico basado en mis parámetros

```python
# Crear una visualización
sns.displot(data=tips, x='total_bill')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill',y='tip')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill',hue='sex')
plt.show()
```

El parámetro "kind" me permite definir cuál gráfico usar en específico: “hist”, “kde”, “ecdf”

```python
sns.displot(data=tips, x='total_bill',hue='sex',kind='kde',legend=True,palette='dark',alpha=.25)
plt.show()
```

```python
# Crea un gráfico de distribución utilizando Seaborn con la función displot
sns.displot(data=tips, x="total_bill", hue='smoker', kind='kde',legend=False, palette='dark',alpha=.25)

plt.title('Diagrama basico') #Título
plt.xlabel('Total de la cuenta') #Eje X
plt.ylabel('Density - kde') #Eje Y
plt.show()
```
