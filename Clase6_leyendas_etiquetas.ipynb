```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

## Añadir títulos, labels y leyenda

```python
fig, axes = plt.subplots(1, 2, figsize=(5, 5))   # Creamos una figura con 1 fila y 2 columnas de ejes, con tamaño de figura de 5x5

axes[0].plot(x, y, label="$sin(x)$")   # Graficamos en el primer eje con "sin(x)" como label
axes[0].set_title('Relación X - Y')   # Establecemos el título del primer eje
axes[0].set_xlabel('X')   # Etiqueta del eje x en el primer eje
axes[0].set_ylabel('Y')   # Etiqueta del eje y en el primer eje
axes[0].legend()   # Agregamos la leyenda al primer eje

axes[1].plot(y, x)   # Graficamos en el segundo eje
axes[1].set_title('Relación Y - X')   # Establecemos el título del segundo eje
axes[1].set_xlabel('Y')   # Etiqueta del eje x en el segundo eje
axes[1].set_ylabel('X')   # Etiqueta del eje y en el segundo eje

fig.tight_layout()    # Para evitar superposición de elementos
```

```python
plt.plot(x, y, label='sin(x)')   # Graficamos la función y = sin(x)
plt.title('Este es un título')   # Establecemos el título del gráfico
plt.xlabel('x')   # Etiqueta del eje x
plt.ylabel('y')   # Etiqueta del eje y
plt.legend()   # Agregamos la leyenda al gráfico
```
