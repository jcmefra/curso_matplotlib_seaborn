```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

## Añadir títulos, labels y leyenda


| Comando              | Descripción                                                                                     |
|----------------------|-------------------------------------------------------------------------------------------------|
| `ax.set_title()`        | Establece el título de la figura.                                                             |
| `ax.set_xlabel()`       | Establece la etiqueta del eje x.                                                              |
| `ax.set_ylabel()`       | Establece la etiqueta del eje y.                                                              |
| `ax.legend()`           | Agrega una leyenda a la figura.                                                                |
| `ax.grid()`             | Muestra una cuadrícula en el gráfico.                                                         |
| `ax.set_xlim()`         | Establece los límites del eje x.                                                               |
| `ax.set_ylim()`         | Establece los límites del eje y.                                                               |

```python
fig, axes = plt.subplots(1, 2, figsize=(5, 5))   # Creamos una figura con 1 fila y 2 columnas de ejes, con tamaño de figura de 5x5

axes[0].plot(x, y, label="$sin(x)$")   # Graficamos en el primer eje con "sin(x)" como label
axes[0].set_title('Relación X - Y')   # Establecemos el título del primer eje
axes[0].set_xlabel('X')   # Etiqueta del eje x en el primer eje
axes[0].set_ylabel('Y')   # Etiqueta del eje y en el primer eje
axes[0].legend()   # Agregamos la leyenda al primer eje

axes[1].plot(y, x)   # Graficamos en el segundo eje
axes[1].set_title('Relación Y - X')   # Establecemos el título del segundo eje
axes[1].set_xlabel('Y')   # Etiqueta del eje x en el segundo eje
axes[1].set_ylabel('X')   # Etiqueta del eje y en el segundo eje

fig.tight_layout()    # Para evitar superposición de elementos
```

```python
plt.plot(x, y, label='sin(x)')   # Graficamos la función y = sin(x)
plt.title('Este es un título')   # Establecemos el título del gráfico
plt.xlabel('x')   # Etiqueta del eje x
plt.ylabel('y')   # Etiqueta del eje y
plt.legend(loc = 'lower left')   # Agregamos la leyenda al gráfico
plt.grid()
```

### Reto: Realizar gráfica personalizada

Armónicos de sin(x)

```python
y = np.zeros(100)
x = np.linspace(0, 20, 100)
n = np.linspace(1,9,9,endpoint=True)

harmonics = []

for idx, element in enumerate(n):
    y = np.sin(x*element)
    harmonics.append(y)

harmonics = np.array(harmonics)
harmonics = np.expand_dims(harmonics, axis=0)
harmonics = harmonics.reshape((3,3,100))
```

```python
nrows = 3
ncols = 3


fig, axes = plt.subplots(nrows, ncols, figsize = (12,6))

count = 0

for idx in range(nrows):
    for jdx in range(ncols):
        count +=1
        axes[idx, jdx].plot(x, harmonics[idx, jdx])
        axes[idx, jdx].set_title(f'Harmonic of $sin(%sx)$' % count)

fig.tight_layout()
```
