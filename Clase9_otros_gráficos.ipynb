```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1,50,100)
```

## Histograma

**plt.hist**

| Parámetro       | Descripción                                                                        |
|-----------------|------------------------------------------------------------------------------------|
| `x`             | Datos para construir el histograma.                                               |
| `bins`          | Número de intervalos (bins) en el histograma.                                     |
| `range`         | Rango de valores a incluir en el histograma.                                     |
| `density`       | Normaliza el histograma para que la integral total sea 1.                         |
| `histtype`      | Tipo de histograma ('bar', 'barstacked', 'step', 'stepfilled').                   |
| `color`         | Color de las barras del histograma.                                               |
| `alpha`         | Transparencia de las barras (entre 0 y 1).                                        |
| `label`         | Etiqueta para la leyenda del gráfico.                                             |
| `edgecolor`     | Color del borde de las barras.                                                    |


```python
plt.hist(data,bins=10,histtype='bar')
plt.show()
```

```python
plt.hist(data,bins=10,histtype='step')
plt.show()
```

## Boxplot

**plt.boxplot**

| Parámetro        | Descripción                                                                   |
|------------------|-------------------------------------------------------------------------------|
| `x`              | Datos para crear los diagramas de caja.                                        |
| `notch`          | Si debe mostrarse un "notch" en la caja (True) o no (False).                  |
| `vert`           | Si la caja debe ser vertical (True) u horizontal (False).                      |
| `patch_artist`   | Si las cajas deben ser dibujadas como objetos artist (True) o no (False).     |
| `widths`         | Ancho relativo de las cajas.                                                  |
| `labels`         | Etiquetas para los diagramas de caja.                                          |
| `showmeans`      | Si deben mostrarse las medias como puntos (True) o no (False).                |
| `shofliers`      | Si deben mostrarse los outliers como puntos (True) o no (False).                |
| `meanline`       | Si debe dibujarse una línea horizontal para indicar la media (True) o no (False).|
| `sym`            | Símbolo para los valores atípicos.                                             |



```python
data = np.append(data,200)
plt.boxplot(data,vert=False,patch_artist=True,notch=True,showfliers=True)
plt.show()
```

## Scatter

**plt.scatter**

| Parámetro       | Descripción                                                                   |
|-----------------|-------------------------------------------------------------------------------|
| `x`             | Coordenadas x de los puntos.                                                 |
| `y`             | Coordenadas y de los puntos.                                                 |
| `s`             | Tamaño de los puntos.                                                        |
| `c`             | Color o secuencia de colores de los puntos.                                  |
| `marker`        | Marca para los puntos (por ejemplo, 'o', 's', 'x').                          |
| `cmap`          | Colormap para mapear valores numéricos a colores.                            |
| `alpha`         | Transparencia de los puntos (entre 0 y 1).                                    |
| `label`         | Etiqueta para la leyenda del gráfico.                                         |
| `edgecolors`    | Color del borde de los puntos.                                               |


```python
N = 50 
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 *  np.random.rand(N)) **2
colors = np.random.rand(N)
```

```python
plt.scatter(x,y,s=area,c=colors,marker='o',alpha=0.5)
plt.show()
```
