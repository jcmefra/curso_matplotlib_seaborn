```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

## plt.subplots()

| Comando              | Descripción                                                                                     |
|----------------------|-------------------------------------------------------------------------------------------------|
| `fig, ax = plt.subplots()` | Crea una nueva figura y un conjunto de ejes (subplots). `fig` representa la figura y `ax` es un array o matriz de ejes (subplots). |
| `ax.plot(x, y)`         | Crea un gráfico en los ejes especificados con los datos de `x` e `y`.                        |
| `ax.set_title()`        | Establece el título de la figura.                                                             |
| `ax.set_xlabel()`       | Establece la etiqueta del eje x.                                                              |
| `ax.set_ylabel()`       | Establece la etiqueta del eje y.                                                              |
| `ax.legend()`           | Agrega una leyenda a la figura.                                                                |
| `ax.grid()`             | Muestra una cuadrícula en el gráfico.                                                         |
| `ax.set_xlim()`         | Establece los límites del eje x.                                                               |
| `ax.set_ylim()`         | Establece los límites del eje y.                                                               |
| `ax.scatter(x, y)`      | Crea un gráfico de dispersión en los ejes con los datos de `x` e `y`.                         |
| `ax.bar(x, height)`     | Crea un gráfico de barras en los ejes con valores de `x` y alturas especificados.             |
| `ax.hist(x, bins)`      | Crea un histograma en los ejes con los datos de `x` y el número de intervalos `bins`.        |
| `ax.pie(x)`             | Crea un gráfico de torta en los ejes con valores proporcionales especificados en `x`.         |


```python
fig, axes = plt.subplots()   # Creamos una figura con un solo eje

axes.plot(x, y, 'b')   # Graficamos en el eje
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)   # Creamos una figura con 1 fila y 2 columnas de ejes

axes[0].plot(x, y, 'b')   # Graficamos en el primer eje
axes[1].plot(y, x, 'r')   # Graficamos en el segundo eje
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)   # Creamos una figura con 1 fila y 2 columnas de ejes

ax1.plot(x, y, 'b')   # Graficamos en el primer eje (ax1)
ax2.plot(y, x, 'r')   # Graficamos en el segundo eje (ax2)
```

```python
fig, axes = plt.subplots(2, 4)   # Creamos una figura con 2 filas y 4 columnas de ejes

axes[0, 0].plot(x, np.cos(x))   # Graficamos en el primer eje de la primera fila
axes[0, 1].plot(x, np.sin(x), 'y')   # Graficamos en el segundo eje de la primera fila con color amarillo ('y')
axes[0, 2].plot(x, np.tan(x), 'r')   # Graficamos en el tercer eje de la primera fila con color rojo ('r')
axes[0, 3].plot(x, np.cos(x)**2)   # Graficamos en el cuarto eje de la primera fila

axes[1, 0].plot(x, np.exp(x), 'g')   # Gráfico de exponencial con color verde
axes[1, 1].plot(x, np.log(x+1), 'm')   # Gráfico de logaritmo con color magenta
axes[1, 2].plot(x, np.sqrt(x), 'b--')   # Gráfico de raíz cuadrada con línea punteada azul
axes[1, 3].plot(x, x**2, 'c-.')   # Gráfico de x^2 con línea guion-punto cian

fig.tight_layout()    # Para evitar superposición de elementos
```

```python
fig, ((ax1, ax2, ax3, ax4), (ax5, ax6, ax7, ax8)) = plt.subplots(2, 4)   # Creamos una figura con 2 filas y 4 columnas de ejes

ax1.plot(x, np.cos(x))   # Graficamos en el primer eje de la primera fila
ax2.plot(x, np.sin(x), 'y')   # Graficamos en el segundo eje de la primera fila con color amarillo ('y')
ax3.plot(x, np.tan(x), 'r')   # Graficamos en el tercer eje de la primera fila con color rojo ('r')
ax4.plot(x, np.cos(x)**2)   # Graficamos en el cuarto eje de la primera fila

ax5.plot(x, np.exp(x), 'g')   # Graficamos en el primer eje de la segunda fila con color verde ('g')
ax6.plot(x, np.log(x+1), 'm')   # Graficamos en el segundo eje de la segunda fila con color magenta ('m')
ax7.plot(x, np.sqrt(x), 'b--')   # Graficamos en el tercer eje de la segunda fila con línea punteada azul ('b--')
ax8.plot(x, x**2, 'c-.')   # Graficamos en el cuarto eje de la segunda fila con línea guion-punto cian ('c-.')

fig.tight_layout()    # Para evitar superposición de elementos
```
