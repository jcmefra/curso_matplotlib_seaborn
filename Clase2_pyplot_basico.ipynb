```python
plt.plot(x,y, 'bx-') #bx- indica hacerlo como markers de color "blue" y en forma de x sobre una línea continua
plt.show()
```

```python
plt.plot(x,y, 'yD:') #yD: indica "yellow" con markers "diamond" y una línea de puntos
plt.show()
```
