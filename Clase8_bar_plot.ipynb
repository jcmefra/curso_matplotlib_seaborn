```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country = ['INDIA','JAPÓN','MÉXICO','COLOMBIA','ALEMANIA']
population = [1408,125,127,51,84]
```

### plt.bar()

| Parámetro      | Descripción                                                                       |
|----------------|-----------------------------------------------------------------------------------|
| `x`            | Posiciones en el eje x para cada barra.                                           |
| `height`       | Alturas de las barras.                                                            |
| `width`        | Ancho de las barras.                                                              |
| `color`        | Color o colores de las barras.                                                    |
| `edgecolor`    | Color del borde de las barras.                                                    |
| `align`        | Alineación de las barras ('center', 'edge').                                      |
| `label`        | Etiqueta para la leyenda del gráfico.                                            |
| `tick_label`   | Etiquetas para los ticks del eje x.                                              |
| `log`          | Escala logarítmica en el eje y (booleano).                                        |
| `alpha`        | Transparencia de las barras (entre 0 y 1).                                       |
| `align`        | Alineación de las barras ('center', 'edge').                                      |
| `orientation`  | Orientación de las barras ('vertical', 'horizontal').                             |
| `hatch`        | Patrón de relleno de las barras ('/', '\\', '|', '-', etc.).                      |



### Reto de la clase: mejorar el bar plot

```python
plt.figure(figsize=(7, 4.5))   # Ajustamos el tamaño de la figura

colors = ['#FF5733', '#3498DB', '#58D68D', '#F39C12', '#9B59B6']   # Colores atractivos para cada país
bars = plt.bar(country, population, width=0.6, color=colors)   # Graficamos las barras con los colores definidos

plt.xlabel('Países')   # Etiqueta del eje x
plt.ylabel('Población (millones)')   # Etiqueta del eje y
plt.title('Población de Países')   # Título del gráfico
plt.xticks(rotation=45)   # Rotamos las etiquetas del eje x para mejor legibilidad
plt.yticks(np.arange(0, 1800, 200))   # Ajustamos las marcas en el eje y

# Agregamos los valores de población en las barras
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 20, round(yval, 1), ha='center', color='black', fontweight='bold')

plt.tight_layout()   # Ajustamos los márgenes y espacios para que no se superpongan los elementos
plt.show()   # Mostramos el gráfico de barras
```
