```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
tips = sns.load_dataset("tips")
tips.head(2)
```

### Gráficos de relación

| Tipo de Gráfico      | Función de Seaborn         | Descripción                                                                                          |
|----------------------|----------------------------|------------------------------------------------------------------------------------------------------|
| Scatterplot          | `sns.scatterplot`          | Representa la relación entre dos variables numéricas utilizando puntos en un plano cartesiano.    |
| Lineplot             | `sns.lineplot`             | Muestra la relación entre dos variables numéricas mediante líneas, ideal para datos secuenciales. |
| Gráfico de Correlación| `sns.heatmap`             | Muestra una matriz de correlación con colores para visualizar las relaciones entre variables.      |
| Gráfico de Conjunto   | `sns.jointplot`           | Muestra la relación entre dos variables numéricas con histogramas y un gráfico de dispersión.    |
| Gráfico de Regresión  | `sns.lmplot`              | Muestra la relación entre dos variables numéricas con una línea de regresión.                    |


```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day',style='time',size='size') #definir parámetros de visualización y datos
plt.legend(loc='center',bbox_to_anchor=(1.12,0.5)) #Acomodar la leyenda
plt.show()
```

```python
plt.figure(figsize=(8,8))
markers = {"Lunch":"D", "Dinner":"s"} #Definir marcadores en un diccionario
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day',style='time',size='size',markers=markers)
plt.legend(loc='center',bbox_to_anchor=(1.12,0.5))
plt.show()
```

```python
sns.lineplot(data=tips,x='total_bill',y='tip',hue='day',style='time',size='size')
plt.legend(loc='center',bbox_to_anchor=(1.12,0.5))
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
plt.title("Regresión Lineal entre Total Bill y Tip")
plt.show()
```

## Al igual que con displot y catplot, se puede usar sns.relplot()

| Parámetro       | Descripción                                                                                         |
|-----------------|-----------------------------------------------------------------------------------------------------|
| `data`          | Conjunto de datos (DataFrame) para la visualización.                                              |
| `x`             | Variable en el eje x para los datos.                                                              |
| `y`             | Variable en el eje y para los datos.                                                              |
| `hue`           | Variable categórica para distinguir elementos por color.                                           |
| `style`         | Variable categórica para distinguir elementos por estilo (por ejemplo, estilo de línea).          |
| `size`          | Variable categórica para distinguir elementos por tamaño.                                          |
| `markers`       | Variable categórica para distinguir elementos por marcador.                                        |
| `col`           | Variable categórica para dividir el gráfico en columnas (facetas).                                 |
| `row`           | Variable categórica para dividir el gráfico en filas (facetas).                                    |
| `kind`          | Tipo específico de gráfico (por ejemplo, `'scatter'`, `'line'`, `'scatterplot'`).                  |
| `height` y `aspect` | Altura del gráfico y relación de aspecto de las facetas.                                          |
| `col_order`     | Orden de las columnas en las facetas.                                                              |
| `row_order`     | Orden de las filas en las facetas.                                                                 |
| `legend`        | Mostrar leyenda (generalmente `True` o `False`).                                                   |
| `xlabel` y `ylabel` | Etiquetas para los ejes x e y respectivamente.                                                   |
| `title`         | Título del gráfico.                                                                                |
| `xlim` y `ylim` | Límites para los ejes x e y respectivamente.                                                      |
| `xticks` y `yticks` | Ubicaciones de las marcas en los ejes x e y respectivamente.                                   |
| `xticklabels` y `yticklabels` | Etiquetas de las marcas en los ejes x e y respectivamente.                                   |


```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='day',style='time',size='size',col='time')
plt.show()
```
